```python
result = []

for value in ords_prods_merged['order_date_of_week'] :
    if value == 0 :
        result.append('Busiest day')
    elif value == 4 :
        result.append('Least day')
    else :
        result.append('Regularly day')
```

```python
ords_prods_merged['busiest day'] = result
```

```python
ords_prods_merged['busiest day'].value_counts(dropna = False)
```

```python
len(result)
```

```python
result2 = []

for value in ords_prods_merged['order_date_of_week'] :
    if value in [0,1] :
        result2.append('Busiest days')
    elif value in [4,3] :
        result2.append('Slowest days')
    else :
        result2.append('Regularly days')
```

```python
len(result2)
```

```python
ords_prods_merged['busiest days'] = result2
```

```python
result3 = []

for value in ords_prods_merged['order_hour_of_day']:
    if value>=9 and value<=16:
        result3.append("Most orders")
    elif value<=6:
        result3.append("Fewest orders")
    else:
        result3.append("Average orders")
```

```python
ords_prods_merged['busiest_period_of_day'] = result3
```
